```python
# Generating the values we are using for the following plot:
housing['ocean_proximity'].value_counts()
```

```python
# Bar chart for those values here:
housing['ocean_proximity'].value_counts().plot(kind = 'bar')
plt.title("ocean_proximity")
plt.xlabel("ocen_proximity")
plt.ylabel("number_of_house");
```

```python
# Generating the values for the following plot using groupby:
housing.groupby('median_house_value')['median_income'].mean().sort_values(ascending=False)
```

```python
# Plotting those values here with bar chart:
housing.groupby('total_bedrooms')['median_income'].mean().sort_values(ascending=True).plot(kind = 'bar',
                                                                                                  color = 'black')
plt.title("relation_between_price_and_ocean_proxility")
plt.xlabel("lcoation")
plt.ylabel("price");
```
