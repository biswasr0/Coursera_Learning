```python
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
#from docx import Document
cv= open('/Users/rbiswa03/Desktop/CV, Cover Letter, Application/Rajesh_K_Biswas_18sep19.txt', 'r').read()
# instantiate a word cloud object
cv_wc= WordCloud(background_color='red',max_words=10000,stopwords=stopwords)
# generate the word cloud
cv_wc.generate(cv)
# display the word cloud
plt.imshow(cv_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# open the file and read it into a variable alice_novel
alice_novel = open('/Users/rbiswa03/Desktop/CV, Cover Letter, Application/Rajesh_Kumar_Biswas_CV.txt', 'r').read()
    
print ('File downloaded and saved!')
stopwords = set(STOPWORDS)
# instantiate a word cloud object
alice_wc = WordCloud(
    background_color='cyan',
    max_words=100000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel)
# display the word cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
